#Assignment 1 Data Visualization with Python
#Question 1 plot a line plot between a and b

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = np.arange(1,11)
b = np.arange(4,14)
plt.plot(a,b)
plt.title('Line-Plot')
plt.xlabel('a value')
plt.ylabel('b value')
plt.show()
```

#Question 2 Plot a line plot showing the sales trend in company 1 and 2:

```python
days = [1,2,3,4,5,6,7]
sales_1 = [200, 150, 130, 280, 601, 254,236]
sales_2 = [150, 365, 378, 698, 420, 540, 351]
```

```python
plt.plot(days, sales_1, color='blue', linewidth=3)
plt.plot(days, sales_2, color='red', linewidth=3)
plt.title('sales trends')
plt.xlabel('days')
plt.ylabel('sales')
plt.show()
```

#Question 3 Create a 3 by 3 subplots:
#multiple plots


```python
x = [1,2,3,4]
y1 = [2,3,4,5]
y2 = [1,4,9,16]
y3 = [14, 30, 60, 25]
y4 = [43,56,78,98]
```

```python
plt.figure(figsize = (8, 8))
plt.subplot(3,3,1)
plt.plot(x,y1,color='red')
plt.title('subplot-1')

plt.subplot(3,3,2)
plt.plot(x,y2,color='blue')
plt.title('subplot-2')

plt.subplot(3,3,3)
plt.plot(x,y3,color='yellow')
plt.title('subplot-3')

plt.subplot(3,3,4)
plt.plot(x,y4,color='green')
plt.title('subplot-4')

plt.show()
```
